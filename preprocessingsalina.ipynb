```python
def load_Data():
    data1 = sio.loadmat('Salinas_corrected.mat')['salinas_corrected']
    labels1 = sio.loadmat('Salinas_gt.mat')['salinas_gt']
    return data1,labels1
data1,labels1=load_Data()
```

```python
def loadData(name):
    if name=='SA':
        data = sio.loadmat('Salinas_corrected.mat')['salinas_corrected']
        labels = sio.loadmat('Salinas_gt.mat')['salinas_gt']
    
    
    return data,labels
data,labels=loadData(dataset)
```

```python


fig = plt.figure(figsize = (7, 7))

ax = fig.subplots(2, 5)

for i in range(2):
  for j in range(5):
    c = np.random.randint(103)
    ax[i][j].imshow(data[:, :, c], cmap='nipy_spectral')
    ax[i][j].axis('off')
    ax[i][j].title.set_text(f"Band - {c}")
    c+=1

fig1 = plt.gcf()
plt.show()
plt.draw()
fig1.savefig('bandvisualization_salina', dpi=100)
```
